# CSV Data Load And Visualizing Pearson Correlation Matrix

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

file_path = "E:/Research data/MS/All-based-Sensory.csv"
data = pd.read_csv(file_path)

data = data.drop(columns=['Image', 'FilledArea'])

pearson_corr = data.corr(method='pearson')
print(pearson_corr)

plt.figure(figsize=(20, 16))
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Pearson Correlation Matrix')
plt.show()
```

# Visualizing Spearman's Rank Correlation Matrix

```python
spearman_corr = data.corr(method='spearman')
print(spearman_corr)

plt.figure(figsize=(20, 16))
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Spearman\'s Rank Correlation Matrix')
plt.show()
```

# Visualizing Kendall's Tau Correlation Matrix

```python
kendall_corr = data.corr(method='kendall')
print(kendall_corr)

plt.figure(figsize=(20, 16))
sns.heatmap(kendall_corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Kendall\'s Tau Correlation Matrix')
plt.show()
```
